```python
import requests

import numpy as np
import pandas as pd
# import ace_tools as tools
```

```python
# This function takes a city and country as input and uses the OpenCage Geocoding API to retrieve the latitude and longitude of the location.
def get_geospatial_info(city, country="Ecuador"):
    # Construct the API request URL with the city, country, and API key
    url = f"https://api.opencagedata.com/geocode/v1/json?q={city},{country}&key={API_KEY}"
    
    # Make an HTTP GET request to the API
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        # If results are available, extract the geometry (latitude and longitude)
        if data['results']:
            location = data['results'][0]['geometry']
            return location['lat'], location['lng']
    # Return None for both latitude and longitude if no data is available
    return None, None
```
